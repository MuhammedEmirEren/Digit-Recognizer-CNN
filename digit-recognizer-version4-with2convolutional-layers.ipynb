```python
training_data = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
training_data.head()
```

```python
training_features = training_data.copy()
training_labels = training_features.pop('label')
training_features = np.array(training_features)
training_features
```

```python
training_labels
```

```python
training_features = training_features.reshape(training_features.shape[0], 28, 28, 1)
training_features = training_features / 255
```

```python
def convolutionalModel():  
    model = tf.keras.models.Sequential(
        [
            tf.keras.Input(shape=training_features[0].shape),
            tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
            tf.keras.layers.MaxPooling2D(2, 2),
            tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
            tf.keras.layers.MaxPooling2D(2, 2),
            
            tf.keras.layers.Flatten(),
            tf.keras.layers.Dense(128, activation='relu'),
            tf.keras.layers.Dense(10, activation='softmax')
        ])
    model.compile(
		optimizer='adam',
		loss='sparse_categorical_crossentropy',
		metrics=['accuracy']
	)
    return model
```

```python
model = convolutionalModel()
```

```python
training_history = model.fit(training_features, training_labels, epochs=10)
```

```python
testing_data = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
testing_features = testing_data.copy()
testing_features = np.array(testing_features)
testing_features = testing_features.reshape(testing_features.shape[0], 28, 28, 1)
testing_features = testing_features / 255.0
```

```python
predictions = model.predict(testing_features)
```

```python
def create_and_save_predictions_csv(predictions, output_filename="/kaggle/working/submission.csv"):
    """
    Creates a pandas DataFrame from predictions and saves it to a CSV file.

    Args:
        predictions (list, dict, or numpy.ndarray): A list of prediction values,
                                                    a dictionary where keys are IDs and values are predictions,
                                                    or a NumPy array of prediction values.
                                                    If a 2D NumPy array (e.g., probabilities for multiple classes),
                                                    it will take the argmax along axis 1 to get the predicted class.
        output_filename (str): The name of the CSV file to save the predictions to.
    """
    processed_predictions = None
    df = None

    if isinstance(predictions, np.ndarray):
        if predictions.ndim == 2:
            processed_predictions = np.argmax(predictions, axis=1).tolist()
            print("Detected 2D NumPy array, applying argmax to get predicted class labels.")
        elif predictions.ndim == 1:
            processed_predictions = predictions.tolist()
            print("Detected 1D NumPy array.")
        else:
            print("Error: NumPy array must be 1D or 2D.")
            return
    elif isinstance(predictions, list):
        processed_predictions = predictions
    elif isinstance(predictions, dict):
        # If predictions is a dictionary, assume keys are IDs and values are targets.
        # The user has specified 'ImageId' and 'Label' for the output format.
        # Assuming the dictionary keys are the correct ImageId values.
        df = pd.DataFrame(list(predictions.items()), columns=['ImageId', 'Label'])
    else:
        print("Error: predictions must be a list, dictionary, or a NumPy array.")
        return

    if processed_predictions is not None:
        # If predictions were processed into a list (from 1D/2D array or original list)
        # Assume an ImageId column needs to be generated, starting from 1.
        # The user has specified 'ImageId' and 'Label' for the output format.
        data = {'ImageId': range(1, len(processed_predictions) + 1), 'Label': processed_predictions}
        df = pd.DataFrame(data)

    expected_columns = ['ImageId', 'Label']
    df = df[expected_columns]

    df.to_csv(output_filename, index=False)
    print(f"Predictions saved to {output_filename}")
```
